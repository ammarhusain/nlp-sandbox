```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import os
import yaml
import time
import random
import numpy as np
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils import data
from tensorboardX import SummaryWriter

# Datasets
from data_loader.mnist import MNISTDataLoader
from data_loader.fashion_mnist import FashionMNISTDataLoader

# Models
from models.gan import ACDiscriminatorNet, ACGeneratorNet
# Loss
from loss.cross_entropy_2d import CrossEntropy2d
# Metrics
from utils.logger import Logger

# https://machinelearningmastery.com/how-to-develop-an-auxiliary-classifier-gan-ac-gan-from-scratch-with-keras/#:~:text=The%20Auxiliary%20Classifier%20GAN%2C%20or,than%20receive%20it%20as%20input.
  
# Some convenience functions
def img_to_vector(img):
  return img.view(img.size(0), 28*28) # [N, X=784]
def vector_to_img(vec):
  return vec.view(vec.size(0), 1, 28, 28) # [N, C=1, H=28, W=28]
def noise(batch_sz, dim):
  return torch.randn(batch_sz, dim)

# Main training function
def train(config, data_set):
  # setup seeds
  torch.manual_seed(config["seed"])
  torch.cuda.manual_seed(config["seed"])
  np.random.seed(config["seed"])
  torch.cuda.empty_cache()
    
  # setup device
  device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
  # setup dataloader
  train_data_set = data_set(split="training")
  train_data = data.DataLoader(train_data_set, batch_size=config["batch_sz"],\
                                num_workers=config["num_workers"], shuffle=True)
  num_batches = len(train_data)
  
  # setup models
  discriminator = ACDiscriminatorNet().to(device)
  generator = ACGeneratorNet(config["noise_dim"]).to(device)
    
  # setup optimizer
  d_optimizer = torch.optim.Adam(discriminator.parameters(), lr=0.0002)
  g_optimizer = torch.optim.Adam(generator.parameters(), lr=0.0002)
  # setup loss
  loss_fn_bce = nn.BCEWithLogitsLoss()
  loss_fn_cce = nn.CrossEntropyLoss()
  
  # setup tensorboard writer & checkpoint dir 
  uniq_name = f"{train_data_set.name}_VGAN_{config.get('exp_name', '')}"
  logger = Logger(uniq_name)
    
  i = 0
  # Load a saved checkpoint
  if config.get("resume_ckpoint") is not None:
    if os.path.isfile(config["resume_ckpoint"]):
      print(f"Loading checkpoint: {config['resume_ckpoint']}")
      ckpoint = torch.load(config["resume_ckpoint"])
      discriminator.load_state_dict(ckpoint["discriminator_state"])
      generator.load_state_dict(ckpoint["generator_state"])
      d_optimizer.load_state_dict(ckpoint["d_optimizer_state"])
      g_optimizer.load_state_dict(ckpoint["g_optimizer_state"])
      i = ckpoint["epoch"]
      print(f"Saved epoch: {i}, loss: {ckpoint['epoch_loss']}, time: {ckpoint['epoch_time']}")
    else:
      raise FileNotFoundError("Unable to load saved checkpoint!, Quitting")
  
  test_noise_vec = noise(16, config["noise_dim"]).to(device)
  test_noise_labels_vec = torch.randint(10, (16, )).to(device) # random integers in [0,10): [batch_sz, 1]

  # Start the epochs. 
  while i < config["epochs"]:
    i += 1
    epoch_loss = 0
    epoch_time = 0
    # Prepare for training
    discriminator.train()
    generator.train()
        
    print(f"Starting Epoch: {i}")
    for b_i, (real_imgs, real_classes) in enumerate(train_data):
      start_ts = time.time()
      real_vecs = img_to_vector(real_imgs).to(device)
      ## 1. Train the discriminator
      d_optimizer.zero_grad()
      # Train on real data
      pred_real, pred_class = discriminator(real_vecs)
      ##(soft target): real_tgt = torch.FloatTensor(pred_real.size(0), 1).uniform_(0.7, 1.2).to(device)
      real_tgt = torch.ones(pred_real.size(0), 1).to(device)
      error_real = loss_fn_bce(pred_real, real_tgt)
      error_real.backward(retain_graph=True)
      # compute categorical cross entropy loss over the class
      error_class = loss_fn_cce(pred_class, real_classes.to(device))
      error_class.backward()
      
      # Train on fake data
      # Generate some fake data using the generator.
      # Detach it here so that gradients are not computed for the 
      # generator when training the discriminator
      noise_vec = noise(real_vecs.size(0), config["noise_dim"]).to(device)
      # produce fake set of labels for embedding.
      noise_labels_vec = torch.randint(10, (real_vecs.size(0), 1)).to(device) # random integers in [0,10): [batch_sz, 1]
      fake_vecs = generator(noise_vec, noise_labels_vec).detach()
      pred_fake, _ = discriminator(fake_vecs) # ignore the pred_class since these are fake images ???
      ##fake_tgt = torch.FloatTensor(pred_fake.size(0), 1).uniform_(0.0, 0.3).to(device)
      fake_tgt = torch.zeros(pred_fake.size(0), 1).to(device)
      error_fake = loss_fn_bce(pred_fake, fake_tgt)
      error_fake.backward()
      # Update weights
      d_optimizer.step()
      
      ## 2. Train the generator
      g_optimizer.zero_grad()
      # Generate some fake data using the generator.
      noise_vec = noise(real_vecs.size(0), config["noise_dim"]).to(device)
      # produce fake set of labels for embedding.
      noise_labels_vec = torch.randint(10, (real_vecs.size(0), )).to(device) # random integers in [0,10): [batch_sz, 1]
      fake_vecs = generator(noise_vec, noise_labels_vec)
      pred_type, pred_class = discriminator(fake_vecs)
      # compute loss wrt to 1s since we want the discriminator 
      # to be fooled to consider these as real
      ##real_tgt = torch.FloatTensor(pred.size(0), 1).uniform_(0.7, 1.2).to(device)
      real_tgt = torch.ones(pred_type.size(0), 1).to(device)
      error = loss_fn_bce(pred_type, real_tgt) 
      error.backward(retain_graph=True)
      # compute categorical cross entropy loss over the class
      error_class = loss_fn_cce(pred_class, noise_labels_vec)
      error_class.backward()
      # Update weights
      g_optimizer.step()
      
            
      ## Display Progress every few batches
      if (b_i) % 100 == 0: 
        test_vecs = generator(test_noise_vec, test_noise_labels_vec)
        test_imgs = vector_to_img(test_vecs).cpu().data
        logger.log_images(test_imgs, i, b_i, num_batches)
        d_error=error_real+error_fake+error_class
        g_error=error
        logger.log_gan_error(d_error, g_error, i, b_i, num_batches)
        logger.display_gan_status(d_error, g_error, pred_real, pred_fake, i, config["epochs"], b_i, num_batches)
        pred_type, pred_class = discriminator(test_vecs)
        print(f"ptype {pred_type.shape}")
        logger.display_classes(test_noise_labels_vec, pred_type.squeeze(), torch.argmax(pred_class, dim=1))
# Setup        
config = {"exp_name": "trial_acgan", "noise_dim": 100, "batch_sz": 256, "epochs": 500,\
          "seed": 1337, "num_workers": 1, "save_ckpoint": True} #"resume_ckpoint": "checkpoints/fcn32s_3.pkl", }

train(config, MNISTDataLoader)
#train(config, FashionMNISTDataLoader)
```
