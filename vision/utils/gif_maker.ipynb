```python
from PIL import Image, ImageDraw
import os
"""
Doc: https://pillow.readthedocs.io/en/stable/handbook/image-file-formats.html
"""
dir = "../checkpoints/anime-faces_DC-GAN_saver_logs/images"
images = []
files = os.listdir(dir)
files.sort(key=lambda x: os.path.getmtime(os.path.join(dir, x)))

for file in files:
  if file.endswith(".png") and file.startswith("_"):
    print (file)
    images.append(Image.open(os.path.join(dir, file)))
    
print(len(images))
images[0].save(os.path.join(dir,"animated.gif"), save_all=True, append_images=images[1:],\
                           optimize=False, loop=0)

```
